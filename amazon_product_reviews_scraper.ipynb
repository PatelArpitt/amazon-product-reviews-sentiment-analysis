```python
# url = input('Please enter the Amazon Product URL here:')
# url = input('Please enter the Amazon Product URL here:')
# url = 'https://www.amazon.ca/Gaming-Keyboard-Backlit-Chromebook-Laptop/product-reviews/B08JPYTS6J/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
url = 'https://www.amazon.ca/Gaming-Keyboard-Backlit-Chromebook-Laptop/product-reviews/B08JPYTS6J/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews&pageNumber=2'

try:
    req = requests.get(url)
except requests.exceptions.Timeout:
    print ('Connection timeout, please retry or wait for some time before retrying.')
except requests.exceptions.TooManyRedirects:
    print ('Too many redirects, please check the URL and try again.')
except requests.exceptions.RequestException as e:
    print ('Fatal error ::: ', e)
    raise SystemExit(e)
else:
    print ('URL is good!')

    name = []
    review_title = []
    review_content = []
    date_info = []
    rating = []
    # data_string = ""

    params = { 'api_key': "57ebc5f03ab018912f6d04cb2b5a0ccc", 'url': url }
    response = requests.get('http://api.scraperapi.com/', params = urllib.parse.urlencode(params))
    soup = BeautifulSoup(response.text, 'html.parser')

    customer_review_lst = soup.find("div", {"class":"a-section a-spacing-none reviews-content a-size-base"})
    customer_review_lst = BeautifulSoup('<br/>'.join([str(tag) for tag in customer_review_lst]), 'html.parser')

    name_lst = customer_review_lst.find_all("span", class_="a-profile-name")
    for name in name_lst:
        name.append(name.get_text())

    title_lst = customer_review_lst.find_all("a", attrs = {"data-hook": "review-title"})
    for title in title_lst:
        review_title.append(title.get_text())

    content_lst = customer_review_lst.find_all("span", attrs = {"data-hook": "review-body"})
    for content in content_lst:
        review_content.append(content.get_text())
        
    date_info_lst = customer_review_lst.find_all("span", attrs = {"data-hook": "review-date"})
    for date in date_info_lst:
        date_info.append(date.get_text())

    rating_lst = customer_review_lst.find_all("i", attrs = {"data-hook": "review-star-rating"})
    for rating in rating_lst:
        rating.append(rating.get_text())

    # Create the dictionary.
    reviews_dict = {'Reviewer Name': name, 'Review Title': review_title, 'Review Content': review_content, 'Date Info': date_info, 'Rating': rating}
    # Print the lengths of each list.  
    print('name', len(name), 'review_title', len(review_title), 'reviews', len(review_content), 'date_info', len(date_info), 'rating', len(rating))
```

```python
# Create a new dataframe.
df = pd.DataFrame.from_dict(reviews_dict, orient='index')
df.head()
```

```python
df.dropna(axis=1, inplace=True)
df.head()
```

```python
# Transpose the dataframe.
prod_reviews = df.T
print(prod_reviews.head(10))
```

```python
# Remove special characters from review text.
prod_reviews['Review Title'] = prod_reviews['Review Title'].str.replace('\n','')
prod_reviews['Review Content'] = prod_reviews['Review Content'].str.replace('\n','')
prod_reviews.head(5)
```

```python
# Convert dataframe to CSV file.
prod_reviews.to_csv('reviews.csv', index=False, header=True)
```
